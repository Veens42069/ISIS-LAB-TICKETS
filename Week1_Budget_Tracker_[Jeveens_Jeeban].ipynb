## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### 💡 Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
# Ask user for monthly income
income = float(input("Enter your monthly income: "))

# Ask user for up to 3 expenses
expense1 = float(input("Enter expense 1: "))
expense2 = float(input("Enter expense 2: "))
expense3 = float(input("Enter expense 3: "))

# Calculate total expenses
total_expenses = expense1 + expense2 + expense3

# Calculate remaining budget
remaining_budget = income - total_expenses

# Display results
print("\n----- Budget Summary -----")
print(f"Total Income: {income:.2f}")
print(f"Total Expenses: {total_expenses:.2f}")
print(f"Remaining Budget: {remaining_budget:.2f}")

# Show warning if budget is negative
if remaining_budget < 0:
    print("⚠ Warning: Your budget is negative! You are overspending.")
else:
    print("✅ Good job! You are within your budget.")
```

```python
# ENHANCED BUDGET TRACKER

from datetime import datetime

# Evidence Package Documentation
evidence_1 = {
    "timestamp": datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
    "context": "I created a budget program with spending categories using AI assistance",
    "ai_tool_used": "ChatGPT",
    "what_worked": "The AI generated structured and readable code",
    "what_i_learned": "I learned how to store related data using dictionaries"
}

# Display evidence
print("----- Evidence Package -----")
for key, value in evidence_1.items():
    print(f"{key}: {value}")

print("\n----- Budget Program -----")

# Ask for monthly income
income = float(input("Enter your monthly income: "))

# Dictionary to store expenses by category
expenses = {}

# Allow user to enter up to 3 expenses
for i in range(1, 4):
    category = input(f"\nEnter category for expense {i} (e.g., Rent, Food, Transport): ")
    amount = float(input(f"Enter amount for {category}: "))
    expenses[category] = amount

# Calculate total expenses
total_expenses = sum(expenses.values())

# Calculate remaining budget
remaining_budget = income - total_expenses

# Display summary
print("\n----- Budget Summary -----")
print(f"Income: {income:.2f}")

print("\nExpenses by Category:")
for category, amount in expenses.items():
    print(f"{category}: {amount:.2f}")

print(f"\nTotal Expenses: {total_expenses:.2f}")
print(f"Remaining Budget: {remaining_budget:.2f}")

# Warning if budget is negative
if remaining_budget < 0:
    print("⚠ Warning: Your budget is negative! You are overspending.")
else:
    print("✅ Good job! You are within your budget.")
```
